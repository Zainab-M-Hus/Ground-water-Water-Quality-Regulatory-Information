```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import string
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("geotracker_sites_2024-11-01.csv", encoding='latin1')
```

```python
df
```

```python
columns_drop = ["BUSINESS_NAME", "GLOBAL_ID", "CLAIM_NO", "CASEWORKER", "LOCAL_AGENCY", "LEAD_AGENCY",
    "RB_CASE_NUMBER", "LOC_CASE_NUMBER", "FILE_LOCATION", "SITE_HISTORY", "STREET_NUMBER", "STREET_NAME","STATE",
    "HOW_DISCOVERED_DESCRIPTION", "STOP_DESCRIPTION", "FACILITY_PROJECT_SUBTYPE", "ZIP", "LATITUDE", "LONGITUDE", "EPA_REGION",
    "COORDINATE_SOURCE", "HOW_DISCOVERED_DESCRIPTION", "DISCHARGE_CAUSE", "STOP_METHOD", "CALWATER_WATERSHED_NAME", "CITY", "DWR_GROUNDWATER_SUBBASIN_NAME" ]
df.drop(columns=columns_drop, inplace=True)
```

```python
def lower(name):
    return name.strip().replace(' ', '_').replace('-', '_').replace('__', '_')
df.columns = [lower(col).replace('(', '').replace(')', '') for col in df.columns]
df
```

```python
date_cols = ["status_date", "no_further_action_date", "leak_reported_date", "begin_date"]
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
df
```

```python
df.select_dtypes(include=['float64']).columns
```

```python
for col in df.select_dtypes(include=['float64']).columns:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip().str.lower()
    df[col].fillna("", inplace=True)
df
```
